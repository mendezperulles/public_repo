**Analysing Titanic Dataset with Copilot**

```python
##import titanic dataset from seaborn
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
##graph the relationship between age and class as a stripplot with a .5 transparency
sns.stripplot(x='class', y='age', data=titanic, alpha=.25)
```

```python
##graph the relationship between survival and sex
sns.barplot(x='sex', y='survived', data=titanic)

```

```python
##graph the percentage of passengers who survived by class
sns.barplot(x='class', y='survived', data=titanic)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'titanic' is your DataFrame
titanic['survived'] = titanic['survived'].replace({0: 'No', 1: 'Yes'})

# Plotting the graph
sns.countplot(x='class', hue='survived', data=titanic)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Survival by Class')
plt.show()
```
