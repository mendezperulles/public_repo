## Visualization & Data Exploration

**Capital Efficiency**

```python
def thousands(x, pos):
    'The two args are the value and tick position'
    return '%1.0fK' % (x * 1e-3)

formatter = ticker.FuncFormatter(thousands)

ax = plt.gca()
ax.yaxis.set_major_formatter(formatter)

# Plot the stacked bar graph
ax.bar(df["Round"], df["LTD_CASH"], color="darkblue", label="Long Term Debt")
ax.bar(df["Round"], df["Sales_From_CS"], bottom=df["LTD_CASH"], color="darkgreen", label="CS Sales")
ax.bar(df["Round"], df["Emergency_Loan"], bottom=df["LTD_CASH"] + df["Sales_From_CS"], color="darkred", label="Emergency Loan")

# Add the legend
plt.legend(loc="upper right")

# Add the title and labels
plt.title("Capital Structure", fontsize=16)
plt.ylabel("USD")
plt.xlabel("")

ax.axhline(0, color="black", linewidth=2)


# Show the plot
plt.show()
```

```python
##create a line graoh using round as x and "ROS" and "CM" as y, and give me a legend


sns.pointplot(data=df, x="Round", y="ROS", color= "DARKred", label = "Return on Sales")
sns.pointplot(data=df, x="Round", y="CM", label="Contribution Margin")
plt.title("Return on Sales & Contribuiton Margin", fontsize=16)
plt.ylabel("")
plt.xlabel("")
plt.legend(loc="upper left")

ax = plt.gca()
ax.axhline(0, color="black", linewidth=2)


plt.show()
```
